```python
df.drop(['SKU TYPE', 'SKU Category','Size (for bags only)*',
       'MPN (UPC)*','Sub category'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df['ID_No']= df['ID_No'].astype(str)
```

```python
customers = df["S.no."].unique().tolist()
len(customers)
```

```python
# shuffle customer ID's
random.shuffle(customers)

# extract 90% of customer ID's
customers_train = [customers[i] for i in range(round(0.9*len(customers)))]

# split data into train and validation set
train_df = df[df['S.no.'].isin(customers_train)]
validation_df = df[~df['S.no.'].isin(customers_train)]
```

```python
from tqdm import tqdm
# list to capture purchase history of the customers
purchases_train = []

# populate the list with the product codes
for i in tqdm(customers_train):
    temp = train_df[train_df["S.no."] == i]["Product Name *"].tolist()
    purchases_train.append(temp)
```

```python
purchases_train
```

```python
purchases_val = []

# populate the list with the product codes
for i in tqdm(validation_df['S.no.'].unique()):
    temp = validation_df[validation_df["S.no."] == i]["Product Name *"].tolist()
    purchases_val.append(temp)
```

```python
purchases_val
```

```python
# train word2vec model
model = Word2Vec(window = 10, sg = 1, hs = 0,
                 negative = 10, # for negative sampling
                 alpha=0.03, min_alpha=0.0007,
                 seed = 14)

model.build_vocab(purchases_train, progress_per=200)

model.train(purchases_train, total_examples = model.corpus_count, 
            epochs=10, report_delay=1)
```

```python
model.init_sims(replace=True)
```

```python
print(model)
```

```python
X = model[model.wv.vocab]

X.shape
```

```python
products = train_df[["ID_No", "Product Name *"]]

# remove duplicates
products.drop_duplicates(inplace=True, subset='ID_No', keep="last")

# create product-ID and product-description dictionary
products_dict = products.groupby('ID_No')['Product Name *'].apply(list).to_dict()
```

```python
products_dict['CIF10000']
```

```python
def similar_products(v, n = 6):
    
    # extract most similar products for the input vector
    ms = model.similar_by_vector(v, topn= n+1)[1:]
    
    # extract name and similarity score of the similar products
    new_ms = []
    for j in ms:
        pair = (products_dict[j[0]][0], j[1])
        new_ms.append(pair)
        
    return new_ms
```

```python
similar_products(model['KATE SMALL IN GRAIN DE POUDRE EMBOSSED LEATHER PRINTED WITH JOUR'])
```

```python
def aggregate_vectors(products):
    product_vec = []
    for i in products:
        try:
            product_vec.append(model[i])
        except KeyError:
            continue
        
    return np.mean(product_vec, axis=0)
```
